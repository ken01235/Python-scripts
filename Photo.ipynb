```python
import cv2
import numpy as np
```

```python
# 讀取兩張圖片
img1 = cv2.imread('../20250331_1.jpg')
img2 = cv2.imread('../Reports/20241226_XPS_FinFET_1T1R/StoMultiFIN_1.25x.jpg')

# 將圖片轉換成 LAB 色彩空間
lab1 = cv2.cvtColor(img1, cv2.COLOR_BGR2LAB)
lab2 = cv2.cvtColor(img2, cv2.COLOR_BGR2LAB)

# 分離通道
l1, a1, b1 = cv2.split(lab1)
l2, a2, b2 = cv2.split(lab2)

# 調整第二張圖的 A、B 通道的平均值，使其接近第一張圖
a2 = a2.astype(np.float32) + (np.mean(a1) - np.mean(a2))
b2 = b2.astype(np.float32) + (np.mean(b1) - np.mean(b2))

# 限制值域並合併回 LAB 圖
a2 = np.clip(a2, 0, 255).astype(np.uint8)
b2 = np.clip(b2, 0, 255).astype(np.uint8)
corrected_lab2 = cv2.merge([l2, a2, b2])

# 轉回 BGR
corrected_img2 = cv2.cvtColor(corrected_lab2, cv2.COLOR_LAB2BGR)

cv2.imwrite('corrected_image2.jpg', corrected_img2)
```
