```python
import os
import shutil
```

```python
# 設定來源資料夾與目標資料夾
source_folder = r"..\raw data"
target_folder = r".\data"

# 確保目標資料夾存在
os.makedirs(target_folder, exist_ok=True)

# 遍歷所有子資料夾，尋找a.txt檔案
for root, dirs, files in os.walk(source_folder):
    if 'for_origin.xlsx' in files:
        # 取得相對路徑作為新檔名依據
        relative_path = os.path.relpath(root, source_folder)
        # 將路徑中的斜線轉為底線，以避免檔名無法使用
        safe_name = relative_path.replace(os.sep, '_')
        new_filename = f'{safe_name}_data.xlsx'

        # 原始檔案路徑
        source_file = os.path.join(root, 'for_origin.xlsx')
        # 目標檔案路徑
        target_file = os.path.join(target_folder, new_filename)

        # 複製檔案
        shutil.copy2(source_file, target_file)
        print(f'複製 {source_file} 至 {target_file}')
```
